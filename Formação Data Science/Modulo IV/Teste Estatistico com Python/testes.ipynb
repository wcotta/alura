Testes Estatísticos

```python
import pandas as pd
import statsmodels as sm
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tmdb = pd.read_csv('../Arquivos/tmdb_5000_movies.csv')
tmdb.head()
```

```python
tmdb.describe()
```

```python
plt.figure(figsize=(5,5))
sns.set_style('darkgrid')
ax = sns.displot(tmdb.vote_average, kde = True)
ax.set(xlabel = 'Nota Média', ylabel = "Frequência",
       title = "Média de votos em filmes no TMDB 5000")
```

```python
ax = sns.boxplot(x = tmdb.vote_average)
ax.set(xlabel = 'Nota média do Filme', title = "Distribuição de nota média dos filmes no TMDB 5000")
```

```python
tmdb_maior_que_10 = tmdb.query('vote_count > 10')
```

```python
sns.set_style('darkgrid')
ax = sns.displot(tmdb_maior_que_10.vote_average, kde = True)
ax.set(xlabel = 'Nota Média', ylabel = "Frequência",
       title = "Média de votos em filmes no TMDB 5000")
```

```python
ax = sns.boxplot(x = tmdb_maior_que_10.vote_average)
ax.set(xlabel = 'Nota média do Filme', title = "Distribuição de nota média dos filmes no TMDB 5000 > 10 votos")
```
